```python
# import libraries
import numpy as np;
from bs4 import BeautifulSoup;
import requests;
import datetime;
import pandas as pd;
```

```python


# Function to get the title of the product
def get_title(soup):
    try:
        title = soup.find("span", attrs={'id': "productTitle"}).get_text().strip()
    except AttributeError:
        title = "No title available"
    return title

# Function to get the price of the product
def get_price(soup):
    try:
        price = soup.find("span", attrs={'class': "a-offscreen"}).get_text().strip()
    except AttributeError:
        price = "No price available"
    return price

# Function to get the rating of the product
def get_start(soup):
    try:
        start = soup.find("span", attrs={'class': "a-icon-alt"}).get_text().strip()
    except AttributeError:
        start = "No start available"
    return start
```
